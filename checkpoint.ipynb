Perguntas:
1) Quem são os 50 melhores jogadores em precisão de chute nesta temporada entre aqueles que fizeram pelo menos 10 chutes?
2) Quais jogadores têm o maior número de gols dentre os mais precisos?
3) Quem são os jogadores de maior eficiência?
4) Quem são os jogadores menos eficientes?
5) Existe quem tenha uma alta quantidade ou precisão mas que não consegue marcar muitos gols?

7) Quem são os jogadores mais equilibrados nos dados?

```python
import pandas as pd
import matplotlib.pyplot as plt
from adjustText import adjust_text

df = pd.read_csv('player_shooting_2023_2024.csv')
df = df[df['Sh'] >= 10]
df = df.sort_values(by='SoT%', ascending=False).head(25)
```

```python
plt.figure(figsize=(12, 8))
scatter = plt.scatter(df['Sh'], df['Gls'], s=df['SoT%']*10, c=df['SoT%'], cmap='viridis', alpha=0.6)
plt.colorbar(scatter, label='Precisão do Chute (%)')
plt.title('Chutes x Gols Marcados (Top 25 da Premier League 23/24, mínimo 10 chutes realizados.)')
plt.xlabel('Chutes')
plt.ylabel('Gols')

texts = [plt.text(df['Sh'].iloc[i], df['Gls'].iloc[i], df['Player'].iloc[i], fontsize=7, ha='center', va='center', color='Black') for i in range(len(df))]
adjust_text(texts, arrowprops=dict(arrowstyle='->', color='red'))

plt.grid(True)
plt.show()
```
